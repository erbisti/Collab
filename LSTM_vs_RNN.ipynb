```python
# =========================================================
# Comparação: SimpleRNN vs LSTM em série com memória curta/longa
# - Sinal = seno(baixa f) + seno(alta f)
# - Seletor: 'rnn', 'lstm' ou 'both'
# - Mede MSE de teste, tempo de treino e tempo de predição
# - Plota previsões sobre o trecho final
# =========================================================

# ------------------------------
# Imports
# ------------------------------
import time
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import SimpleRNN, LSTM, Dense
from tensorflow.random import set_seed
from sklearn.metrics import mean_squared_error

# ------------------------------
# Seletor
# ------------------------------
SELECTOR = 'both'   # opções: 'rnn', 'lstm', 'both'

# ------------------------------
# Reprodutibilidade (opcional)
# ------------------------------
np.random.seed(22)
set_seed(22)

# ------------------------------
# Geração de dados
# ------------------------------
N_PONTOS = 1000
X_MAX    = 120.0
x = np.linspace(0, X_MAX, N_PONTOS)

# Componente de baixa f (memória longa) + alta f (memória curta)
AMP_LOW,  W_LOW  = 1.0, 0.15
AMP_HIGH, W_HIGH = 0.4, 1.20
NOISE_STD = 0.00

y = AMP_LOW*np.sin(W_LOW*x) + AMP_HIGH*np.sin(W_HIGH*x)
if NOISE_STD > 0:
    y = y + np.random.normal(scale=NOISE_STD, size=y.shape)

# ------------------------------
# Janela deslizante
# ------------------------------
def create_dataset(seq, n_steps):
    Xs, Ys = [], []
    for i in range(len(seq) - n_steps):
        Xs.append(seq[i:i + n_steps])
        Ys.append(seq[i + n_steps])
    return np.array(Xs), np.array(Ys)

# ------------------------------
# Parâmetros principais
# ------------------------------
n_steps    = 50     # janela temporal (suficiente p/ captar baixa f)
test_len   = 200    # trecho final reservado p/ teste
units_rnn  = 64     # neurônios na SimpleRNN
units_lstm = 64     # neurônios na LSTM
epochs     = 30
batch_sz   = 64

# ------------------------------
# Dataset e split temporal
# ------------------------------
X, Y = create_dataset(y, n_steps)               # (T-n_steps, n_steps)
X = X.reshape((X.shape[0], X.shape[1], 1))      # (amostras, passos, features)
T = len(y)
train_end = T - test_len                        # corte na série original
train_count = train_end - n_steps               # i < train_end - n_steps

assert train_count > 0, "Ajuste test_len/n_steps/N_PONTOS."
X_train, Y_train = X[:train_count], Y[:train_count]
X_test,  Y_test  = X[train_count:], Y[train_count:]   # tamanho esperado = test_len

# ------------------------------
# Fábrica de modelos
# ------------------------------
def build_model(kind, units):
    model = Sequential()
    if kind == 'rnn':
        model.add(SimpleRNN(units, activation='tanh', input_shape=(n_steps, 1)))
    elif kind == 'lstm':
        model.add(LSTM(units, activation='tanh', input_shape=(n_steps, 1)))
    else:
        raise ValueError("kind inválido: use 'rnn' ou 'lstm'")
    model.add(Dense(1))
    model.compile(optimizer='adam', loss='mse')
    return model

# ------------------------------
# Treino + avaliação
# ------------------------------
def train_and_eval(kind, units):
    model = build_model(kind, units)
    print("\n" + "="*60)
    print(f"Modelo: {kind.upper()} (units={units})")
    model.summary()

    t0 = time.perf_counter()
    hist = model.fit(
        X_train, Y_train,
        epochs=epochs,
        batch_size=batch_sz,
        verbose=1
    )
    t_train = time.perf_counter() - t0

    t1 = time.perf_counter()
    pred_test = model.predict(X_test)
    t_pred = time.perf_counter() - t1

    mse = mean_squared_error(Y_test, pred_test.flatten())
    params = model.count_params()

    result = {
        'kind': kind,
        'units': units,
        'mse_test': mse,
        't_train_s': t_train,
        't_pred_s': t_pred,
        'params': params,
        'pred_test': pred_test.flatten()
    }
    return result

# ------------------------------
# Rodar conforme seletor
# ------------------------------
results = []
if SELECTOR in ('rnn', 'both'):
    results.append(train_and_eval('rnn', units_rnn))
if SELECTOR in ('lstm', 'both'):
    results.append(train_and_eval('lstm', units_lstm))

# ------------------------------
# Plot das previsões no trecho final
# ------------------------------
plt.figure(figsize=(12, 6))
plt.plot(range(T), y, label='Real')

for res in results:
    label = f"Pred. ({res['kind'].upper()})  MSE={res['mse_test']:.4e}"
    plt.plot(
        range(train_end, train_end + len(res['pred_test'])),
        res['pred_test'],
        label=label
    )

plt.axvline(train_end, linestyle='--', linewidth=1, label='Início do teste')
plt.xlabel('Índice (tempo discreto)')
plt.ylabel('Valor')
plt.title('Comparação: SimpleRNN vs LSTM — previsão no trecho final')
plt.legend()
plt.tight_layout()
plt.show()

# ------------------------------
# Tabela simples no console
# ------------------------------
print("\nResumo:")
print("Modelo\tUnits\tParams\t\tMSE_teste\tTreino(s)\tPred(s)")
for r in results:
    print(f"{r['kind']}\t{r['units']}\t{r['params']}\t{r['mse_test']:.6e}\t{r['t_train_s']:.2f}\t\t{r['t_pred_s']:.4f}")

# ------------------------------
# Referência rápida de parâmetros usados
# ------------------------------
print("\nParâmetros principais:")
print(f" n_steps={n_steps}, test_len={test_len}, epochs={epochs}, batch_sz={batch_sz}")
print(f" Série: AMP_LOW={AMP_LOW}, W_LOW={W_LOW}; AMP_HIGH={AMP_HIGH}, W_HIGH={W_HIGH}; NOISE_STD={NOISE_STD}")
print(f" Dimensões: X total {X.shape}, Y total {Y.shape}")
print(f" Treino: X_train {X_train.shape}, Y_train {Y_train.shape}")
print(f" Teste : X_test  {X_test.shape},  Y_test  {Y_test.shape}")
```
