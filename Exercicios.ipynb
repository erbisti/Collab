```python
# Importação das bibliotecas necessárias
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Definindo a configuração de exibição do pandas para melhor visualização dos dados
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
# Leitura dos dados do arquivo Excel
arquivo_excel = 'C:\\Users\\erbisti\\Documents\\Planilha_01.xlsx'
df = pd.read_excel(arquivo_excel)

# Exibindo as primeiras linhas do DataFrame para entender a estrutura dos dados
df.head()
```

```python
# Seleção das colunas numéricas
colunas_numericas = ['IDADE', 'NOTA1', 'NOTA2', 'NOTA3']
df_numerico = df[colunas_numericas]

# Cálculo da média para as colunas numéricas
medias = df_numerico.mean()
print("Médias das colunas numéricas:")
print(medias)

```

```python
# Importando bibliotecas necessárias
import pandas as pd
from scipy.stats import wilcoxon


# Filtrando apenas as colunas necessárias
nota1 = df['NOTA1']
nota2 = df['NOTA2']
nota3 = df['NOTA3']

# Teste de Wilcoxon pareado entre NOTA1 e NOTA2
stat1, p_value1 = wilcoxon(nota1, nota2, alternative='less')

# Teste de Wilcoxon pareado entre NOTA2 e NOTA3
stat2, p_value2 = wilcoxon(nota2, nota3, alternative='less')

# Resultados formatados
print("\n========================================")
print("Teste de Wilcoxon - Comparação Pareada")
print("========================================")
print(f"Comparação NOTA1 vs NOTA2")
print(f"Estatística do teste: {stat1:.4f}")
print(f"p-Valor: {p_value1:.4f}")
print("----------------------------------------")
print(f"Comparação NOTA2 vs NOTA3")
print(f"Estatística do teste: {stat2:.4f}")
print(f"p-Valor: {p_value2:.4f}")
print("========================================\n")

# Interpretação dos resultados
if p_value1 < 0.05:
    print("Resultado: Rejeitamos a hipótese nula H0: NOTA1 >= NOTA2")
    print("Conclusão: Há evidências de que NOTA1 < NOTA2")
else:
    print("Resultado: Não rejeitamos a hipótese nula H0: NOTA1 >= NOTA2")
    print("Conclusão: Não há evidências suficientes de que NOTA1 < NOTA2")

if p_value2 < 0.05:
    print("Resultado: Rejeitamos a hipótese nula H0: NOTA2 >= NOTA3")
    print("Conclusão: Há evidências de que NOTA2 < NOTA3")
else:
    print("Resultado: Não rejeitamos a hipótese nula H0: NOTA2 >= NOTA3")
    print("Conclusão: Não há evidências suficientes de que NOTA2 < NOTA3")
```

```python
import pandas as pd
from scipy.stats import chi2_contingency
import seaborn as sns
import matplotlib.pyplot as plt


# Tabela de contingência
contingency_table = pd.crosstab(df['NIVEL'], df['FX_IDADE'])

# Teste Qui-Quadrado
chi2_stat, p_value, dof, expected = chi2_contingency(contingency_table)
print(f"Chi2 Statistic: {chi2_stat}, p-Value: {p_value}, Degrees of Freedom: {dof}")
print("Expected Frequencies:\n", expected)

# Visualização gráfica da tabela de contingência
plt.figure(figsize=(10, 8))
sns.heatmap(contingency_table, annot=True, cmap="YlGnBu", fmt="d")
plt.title('Tabela de Contingência entre NIVEL e FX_IDADE')
plt.xlabel('FX_IDADE')
plt.ylabel('NIVEL')
plt.show()
```

```python
# Importando bibliotecas necessárias
import pandas as pd
from scipy.stats import mannwhitneyu


# Filtrando os dados para as turmas A e B
nota1_turma_a = df[df['TURMA'] == 'A']['NOTA1']
nota1_turma_b = df[df['TURMA'] == 'B']['NOTA1']

# Teste Mann-Whitney
stat, p_value = mannwhitneyu(nota1_turma_a, nota1_turma_b, alternative='two-sided')

# Resultados formatados
print("\n========================================")
print("Teste Mann-Whitney - Comparação entre Turmas A e B")
print("========================================")
print(f"Comparação NOTA1 entre Turma A e Turma B")
print(f"Estatística do teste: {stat:.4f}")
print(f"p-Valor: {p_value:.4f}")
print("========================================\n")

# Interpretação dos resultados
if p_value < 0.05:
    print("Resultado: Rejeitamos a hipótese nula H0: As distribuições de NOTA1 são iguais entre Turma A e Turma B")
    print("Conclusão: Há evidências de que as distribuições de NOTA1 são diferentes entre Turma A e Turma B")
else:
    print("Resultado: Não rejeitamos a hipótese nula H0: As distribuições de NOTA1 são iguais entre Turma A e Turma B")
    print("Conclusão: Não há evidências suficientes de que as distribuições de NOTA1 são diferentes entre Turma A e Turma B")
```
